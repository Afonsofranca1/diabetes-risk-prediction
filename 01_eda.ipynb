```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style='whitegrid')
```

```python
# Load the PIMA Indian Diabetes dataset from online source
url = "https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv"
df = pd.read_csv(url)

# Preview the data
df.head()
```

```python
# Shape and info
print(f"Shape of dataset: {df.shape}")
df.info()
```

```python
df.describe()
```

```python
# Check how many columns have zeros where they shouldn't (e.g., BMI, Glucose)
zero_cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for col in zero_cols:
    print(f"{col} has {(df[col] == 0).sum()} zero entries")
```

```python
df_cleaned = df.copy()
df_cleaned[zero_cols] = df_cleaned[zero_cols].replace(0, np.nan)

# Visualize missing values
sns.heatmap(df_cleaned.isnull(), cbar=False, cmap='viridis')
plt.title("Missing or Zero-like Values in the Dataset")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.countplot(data=df, x='Outcome')
plt.title("Distribution of Diabetes Outcome")
plt.xticks([0, 1], ['Non-Diabetic', 'Diabetic'])
plt.show()
```

```python
features = df.columns[:-1]  # exclude Outcome
df[features].hist(figsize=(12, 10), bins=20)
plt.tight_layout()
plt.show()
```
